```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
fig = plt.figure()
fig.suptitle('Orris\'s title')  # add a centered to the figure
fig, ax_list = plt.subplots(2, 2) # subplots create a figure and a set of subplots
# help(plt.subplot) # add a subplot to the current figure
```

The Axes object contains two or three Axis objects which take care of the data limit

One Axis object can only be in one figure

One Axes object has its own title.

```
import numpy as np
from scipy.stats import norm
x = np.linspace(-5, 5, 100)
y = norm.pdf(x) # pdf of normal distribution. norm.pdf(x) = exp(-x**2 / 2) / sqrt(2 * pi)
plt.figure()
plt.plot(x, y, label='normal distribution') # the label is triggered by plt.legend() and displayed on the top-right corner
plt.legend() # place the legends on the figure
plt.show()
```

```
plt.figure()
plt.plot(x, y)
plt.xlabel('x label') # no need for trigger
plt.ylabel('y label')
plt.figure()
```

```
plt.figure()
plt.plot(x, y)
plt.title('new figure') # no need for trigger
plt.show()
```

## Recommended Coding Style

```
x = np.linspace(-5, 5, 1000)
y = norm(1.9, 0.5).pdf(x) # set mean and std
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## Interactive backend

```
plt.ion()
plt.plot(x, y) # interactive backend directly shows the figure below
```
