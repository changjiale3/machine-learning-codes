# Extract HOG feature using opencv

## 1. param

```python
import cv2
import numpy as np

image = cv2.imread("test.jpg",0)
winSize = (28, 28)
blockSize = (14, 14)
blockStride = (7, 7)
cellSize = (7, 7)
nbins = 9
derivAperture = 1
winSigma = 4.
histogramNormType = 0
L2HysThreshold = 2.0000000000000001e-001
gammaCorrection = 1
nlevels = 64
hog = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nbins,derivAperture,winSigma, histogramNormType,L2HysThreshold,gammaCorrection,nlevels)

img = np.arange(784).reshape(28, 28)
img = img // 4
img = img.astype(np.uint8) # IMPORTANT!!
feature = hog.compute(img)

print(feature.shape)
print(feature.squeeze())
```
